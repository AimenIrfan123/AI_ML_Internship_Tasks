```python
# Load necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the built-in Iris dataset using seaborn
iris = sns.load_dataset('iris')
```

```python
# Show first 5 rows
iris.head()
```

```python
# See how big the data is
print("Shape:", iris.shape)

# See the column names
print("Column Names:", iris.columns)

# Basic description of numbers
iris.describe()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
sns.histplot(iris['petal_length'], kde=True)
plt.title("Petal Length Distribution")
plt.show()
```

```python
sns.boxplot(x='species', y='petal_width', data=iris)
plt.title("Petal Width by Species")
plt.show()
```

✅ Task 1 Summary:
- We used the Iris dataset which contains 150 flower samples.
- It has 4 features: sepal length, sepal width, petal length, petal width.
- We used scatter plots, histograms, and box plots to visualize the data.
- Each species shows different flower part sizes.
