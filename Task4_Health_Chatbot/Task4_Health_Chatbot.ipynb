```python
# Load a free text-generation model
chatbot = pipeline("text2text-generation", model="google/flan-t5-small")
```

```python
def ask_health_bot(question):
    prompt = f"Answer like a friendly health assistant: {question}"
    response = chatbot(prompt, max_length=100)[0]['generated_text']
    return response

# Try it out!
print(ask_health_bot("What causes a sore throat?"))
```

```python
print(ask_health_bot("Is paracetamol safe for children?"))
print(ask_health_bot("What is the normal body temperature?"))
```
